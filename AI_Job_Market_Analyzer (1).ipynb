```python
descriptions = " ".join(df["description"].fillna("").astype(str))
```

```python
skills = [
    "Python",
    "SQL",
    "Excel",
    "Power BI",
    "Tableau",
    "Machine Learning",
    "Data Analysis",
    "Statistics",
    "Pandas",
    "R"
]
```

```python
skill_count = {}

for skill in skills:
    skill_count[skill] = descriptions.lower().count(skill.lower())

skill_count
```

```python
skill_df = pd.DataFrame(
    list(skill_count.items()),
    columns=["Skill", "Count"]
)

skill_df = skill_df.sort_values(
    by="Count",
    ascending=False
)

skill_df
```

```python
skill_df.plot(
    x="Skill",
    y="Count",
    kind="bar"
)

plt.title("Most Demanded Skills")

plt.show()
```
